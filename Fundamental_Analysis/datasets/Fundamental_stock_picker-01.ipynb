## Key metrics:

- Operating Margin — determines the profit a business generates on each dollar of sales after paying for variable costs such as wages and raw materials
- Dividend Yield — shows the annual dividend payout of a firm in relation to its stock price
- Dividend Cover — measures the number of dividend payments a company can pay to its shareholders
- Debt/EBITDA — measures the amount of income generated and available for debt repayment before interest, taxes, depreciation, and amortization costs are taken into consideration
- P/E ratio — measures its current share price relative to its earnings per share (EPS)
- PEG ratio — evaluates a stock's value while also taking the company's anticipated profits growth into account
